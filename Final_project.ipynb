```python
filteredCountry = dataset[dataset["country"].notnull()]
filteredContinent = dataset[dataset["continent"].notnull()]
```

```python
t5country = filteredCountry["country"].value_counts()
t5country.head()
```

top 5 continent with most earthquakes

```python
t5continent = filteredContinent["continent"].value_counts()
t5continent.head()
```

```python
subset1 = dataset[dataset['country'].notnull()]
countryEQ = pd.DataFrame({'country': subset1['country'].value_counts().index, 'earthquakes': subset1['country'].value_counts().values})
top5country = countryEQ.head()
top5country
```

```python
subset2 = dataset[dataset['continent'].notnull()]
continentEQ = pd.DataFrame({'continent': subset2['continent'].value_counts().index, 'earthquakes': subset2['continent'].value_counts().values})
continentEQ
```

```python
top5country.plot.barh(x='country', y='earthquakes', color="green")
plt.xlabel("No. of Earthquakes")
plt.ylabel("Country")
plt.title("Top 5 countries with the most number of Earthquake from 2001â€“2023");
```

```python
continentEQ.plot.barh(x='continent', y='earthquakes', color="red")
plt.xlabel("No. of Earthquakes")
plt.ylabel("Countries")
plt.title("Continents with the most number of Earthquake from 2001â€“2023");
```

#2 Highest magnitude

```python
title = list(dataset["title"])
magnitude = list(dataset["magnitude"])
title_dict = {key: value for key, value in zip(title, magnitude)}
highest_magnitude = {key: value for key, value in title_dict.items() if value == max(title_dict.values())}
print("Highest magnitude: ", highest_magnitude)
```

```python
subset3 = dataset[['magnitude','mmi']]
subset3
```

```python
subset3.plot.scatter(x="mmi", y="magnitude", color="red")
plt.xlabel("Intensity")
plt.ylabel("Magnitude")
plt.title("Intensity vs Magnitude");
```

```python
dmy =  dataset[['day', 'month', 'year']] = dataset['date_time'].str.split(pat='-', expand=True)
year =  dataset[['year', 'time']] = dataset['year'].str.split(pat=' ', expand=True)
dataset
```

```python
yearpattern = pd.DataFrame({'year': dataset['year'].value_counts().sort_index().index,'No. of earthquakes': dataset['year'].value_counts().sort_index().values})
yearpattern.plot.line(x='year', y='No. of earthquakes', color="red", marker="o")
yearpattern
plt.xlabel("Year")
plt.ylabel("No. of Earthquakes")
plt.title("Earthquake pattern per year from 2001-2022")
```

```python
# cindex=['january','february','march','april','may','june','july','august','september','october','november','december']
monthpattern = pd.DataFrame({'month': dataset['month'].value_counts().sort_index().index,'No. of earthquakes': dataset['month'].value_counts().sort_index().values})
monthpattern.plot.line(x='month', y='No. of earthquakes', color="red", marker="o")
plt.xlabel("Month")
plt.ylabel("No. of Earthquakes")
plt.title("Earthquake pattern per month from 2001-2022")
```

```python
indonesia = dataset.loc[dataset["country"] == "Indonesia"]
indopattern = pd.DataFrame({'year': indonesia['year'].value_counts().sort_index().index,'No. of earthquakes': indonesia['year'].value_counts().sort_index().values})
indopattern
indopattern.plot.line(x='year', y='No. of earthquakes', color="red", marker="o")
plt.xlabel("Year")
plt.ylabel("No. of Earthquakes")
plt.title("Indonesia major earthquake pattern per year")
```

```python
asia = dataset.loc[dataset["continent"] == "Asia"]
asiapattern = pd.DataFrame({'year': asia['year'].value_counts().sort_index().index,'No. of earthquakes': asia['year'].value_counts().sort_index().values})
asiapattern
asiapattern.plot.line(x='year', y='No. of earthquakes', color="red", marker="o")
plt.xlabel("Year")
plt.ylabel("No. of Earthquakes")
plt.title("Asia earthquake pattern per year")
```

```python
fig = plt.figure(figsize=(12,9))

m = Basemap(projection = 'mill', llcrnrlat = -90, urcrnrlat = 90, llcrnrlon = -180, urcrnrlon = 180, resolution = 'i')
m.drawcoastlines(linewidth=1.0, linestyle='solid', color='black')
m.drawcountries(linewidth=1.0, linestyle='solid', color='k')
m.fillcontinents(color='green',lake_color='blue', alpha=0.9)
m.drawmapboundary(fill_color='blue')

lat = dataset['latitude'].tolist()
lon = dataset['longitude'].tolist()

m.scatter(lon, lat, latlon = True, s = 10, c = 'red', marker = 'o', alpha = 1)

plt.show()
```
